### Успешна или убыточна акция?

```python
# Проанализируем цены на рынке наиболее известных фастфуд ресторанов в Санкт-Петербурге
# Ценники на картошку и минимальная стоимость съедобных блюд 
# (возьмём в расчёт, что покупатели не приобретали отдельно соусы, стикеры, салфетки и т.п.)

dodo_pizza_potato = 199
dodo_pizza_min_bon_aqua = 75

kfc_potato = 59
kfc_min_icecream = 39

bk_potato = 130
bk_min_cheese_bites = 69.99

black_star_potato = 150
black_star_min_bon_aqua = 99

dominos_potato = 189
dominos_min_drinks = 99
```

```python
# Найдём отношение минимальной цены на блюдо в ресторанах к цене картошки

min_ratio = (dodo_pizza_min_bon_aqua/dodo_pizza_potato + \
             kfc_min_icecream/kfc_potato + \
             bk_min_cheese_bites/bk_potato + \
             black_star_min_bon_aqua/black_star_potato + \
             dominos_min_drinks/dominos_potato)/5
print(min_ratio)
```

```python
# Определим преполагаемую цену минимальной по цене позиции в "Да-да пицце", 
# исходя из полученного ранее отношения

da_da_pizza_potato = 100
da_da_pizza_estimated_min = round(da_da_pizza_potato*min_ratio)
print(da_da_pizza_estimated_min)
```

```python
# В условии сказано, что ресторан посетило 200 человек.
# Предположим, что посетили = что-то купили
# 20 человек потратили 400+
# 100 человек воспользовались предложением и взяли свой бесплатный картофель => 80 человек потратили от 200 до 400 рублей
# Примем, что если человек не взял бесплатный картофель, значит потратил меньше 200 рублей = > 100 человек потратили до 200 рублей

customers_under_200 = 100
customers_between_200_400 = 80
customers_plus_400 = 20
```

```python
# Сумма потраченных денег клиентов, потративших до 200 рублей

mean_check_under_200 = (200 + da_da_pizza_estimated_min)/2
sum_under_200 = mean_check_under_200*customers_under_200
```

```python
# Сумма потраченных денег клиентов, потративших до от 200 до 400 рублей

mean_check_between_200_400 = (200 + 400)/2
sum_between_200_400 = mean_check_between_200_400*customers_between_200_400
```

```python
# Сумма потраченных денег клиентов, потративших до от более 400 рублей
# Предположим негативный сценарий 
# Клиенты потратившие больше 400 рублей не тратили больше 500 рублей

mean_check_plus_400 = (400 + 500)/2
sum_plus_400 = mean_check_plus_400*customers_plus_400
```

```python
# Общая сумма покупок и стоимость подаренного картофеля

sum_purchases = sum_under_200 + sum_between_200_400 + sum_plus_400
sum_potato = da_da_pizza_potato*100
```

```python
# Рассчитаем при каком фудкосте(включает затраты на сырье, постоянные и дополнительные издрежки) блюд мы выйдем в ноль
# Не будем учитывать предполагаемую прибыль картофеля, которую компания могла получить при его реализации
# Примем, что фудкост для всех блюд одинаков (фудкост = (себестоимость продуктов / сумму выручки)*100)
# x - фудкост
# 1-x - часть стоимости блюда, представляющая чистую прибыль

# (1-x)*sum_purchases - x*sum_potato = 0
# sum_purchases - x*sum_purchases - x*sum_potato = 0
# sum_purchases = x*(sum_purchases + sum_potato)

x = sum_purchases / (sum_purchases + sum_potato)

print(f'''Итог: Фудкост, при котором бизнес не получит прибыли, но и не потеряет её равен {round(x*100,2)}%.''')
```

```python
# Опираясь на данные из открытых источников рассмотрим оптимальную цену фудкоста: 
# https://rkeeper.ru/blog/zachem-schitat-natsenku-i-fudkost/   20-30%
# https://blog.quickresto.ru/chto-takoe-fudkost                25–40%
# https://open-s.info/company/news/kak-schitat-fudkost/        20-30%

# Предположим, что компания "Да-да пицца" не первый день на рынке, налажены цепочки поставок, 
# может позволить себе мобильное приложение
# Примем фудкост в районе 50%, условно включив туда ещё и затраты на маркетинговую компанию
# Рассчитаем успешность акции в денежном выражении

foodcost = 0.5

revenue_per_company = (1-foodcost)*sum_purchases - foodcost*sum_potato

print(f'''Итог: При фудкосте, равном {round(foodcost*100,2)}%, прибыль составит {revenue_per_company} руб.''')
```

### Вывод:
При проведении маркетинговой компании, компания получила прибыль.

Также стоит учесть, что при проведении расчётов мы не учли низкую себестоимость картофеля, а также постэффект от данного мероприятия, который с большей долей вероятности будет иметь положительное влияние на бизнес (возврат ушедших клиентов, увеличение общего впечатления от компании "Да-да пицца", настраивании более глубокой связи между клиентом и безнесом, часть возвратившихся клиентов превратятся в постоянных).
